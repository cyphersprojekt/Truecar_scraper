```python
#Crear listas vacías dentro de las que vamos a añadir la información que deseamos obtener del website
marca_ny = []
modelo_ny = []
año_ny = []
millaje_ny = []
precio_ny = []
```

```python
#Loop que modifica la clave 'page' de nuestros parámetros para obtener la cantidad de resultados que deseamos
#Todas las variables y listas ahora deben estar dentro de este loop
#El funcionamiento en sí es un copy-paste del singular
for i in range(1,170): 
    params = (
    ('city', 'new-york'),
    ('collapse', 'true'),
    ('fallback', 'true'),
    ('include_incentives', 'true'),
    ('include_targeted_incentives', 'true'),
    ('new_or_used', 'u'),
    ('page', str(i)),
    ('per_page', '30'),
    ('postal_code', '10001'),
    ('search_event', 'true'),
    ('sort^/[^/]', 'best_match'),
    ('sponsored', 'true'),
    ('state', 'ny'),
    )  
    
    response = requests.get('https://www.truecar.com/abp/api/vehicles/used/listings', headers=headers, params=params)
    jsonresponse = response.json()
    listings = jsonresponse['listings']
    
    for i in range(len(listings)):
        marca_ny.append(listings[i]['vehicle']['make'])
        modelo_ny.append(listings[i]['vehicle']['model'])
        año_ny.append(listings[i]['vehicle']['year'])
        millaje_ny.append(listings[i]['vehicle']['mileage'])
        precio_ny.append(listings[i]['vehicle']['list_price'])
```

```python
#Crear nuestro dataframe
nydf5k = pd.DataFrame({'Marca_ny':marca_ny, 'Modelo_ny':modelo_ny, 'Año_ny':año_ny, 'Millaje_ny':millaje_ny, 'Precio_ny':precio_ny })
```

```python
#Exportar nuestro dataframe
nydf5k.to_excel('nydf5k.xlsx', index=False)
```
