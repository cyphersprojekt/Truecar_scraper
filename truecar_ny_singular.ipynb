```python
#Crear listas vacías dentro de las que vamos a añadir la información que deseamos obtener del website
marca_ny = []
modelo_ny = []
año_ny = []
millaje_ny = []
precio_ny = []
```

```python
#Loop que va desde 0 hasta el número de items contenidos en la variable 'listings', añadiendo por orden los datos
#que deseamos a nuestras listas
for i in range(len(listings)):
    marca_ny.append(listings[i]['vehicle']['make'])
    modelo_ny.append(listings[i]['vehicle']['model'])
    año_ny.append(listings[i]['vehicle']['year'])
    millaje_ny.append(listings[i]['vehicle']['mileage'])
    precio_ny.append(listings[i]['vehicle']['list_price'])
```

```python
#Crear un DataFrame que presenta nuestra información de la forma en que la deseamos
nydf = pd.DataFrame({'Marca_ny':marca_ny, 'Modelo_ny':modelo_ny, 'Año_ny':año_ny, 'Millaje_ny':millaje_ny, 'Precio_ny':precio_ny })
```

```python
#Exportar nuestro DataFrame hacia un archivo excel
nydf.to_excel('nydf.xlsx', index=False)
```
