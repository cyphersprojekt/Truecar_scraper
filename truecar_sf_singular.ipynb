```python
#Asignar una variable al objeto JSON de la request para facilitar su acceso
jsonresponse = response.json()
```

```python
#Asignar una variable a la clave 'listings' del diccionario para facilitar su acceso
listings = jsonresponse['listings']
```

```python
#Crear listas vacías dentro de las que vamos a añadir la información que deseamos obtener del website
marca_sf = []
modelo_sf = []
año_sf = []
millaje_sf = []
precio_sf = []
```

```python
#Loop que va desde 0 hasta el número de items contenidos en la variable 'listings', añadiendo por orden los datos
#que deseamos a nuestras listas
for i in range(len(listings)):
    marca_sf.append(listings[i]['vehicle']['make'])
    modelo_sf.append(listings[i]['vehicle']['model'])
    año_sf.append(listings[i]['vehicle']['year'])
    millaje_sf.append(listings[i]['vehicle']['mileage'])
    precio_sf.append(listings[i]['vehicle']['list_price'])
```

```python
sfdf = pd.DataFrame({'Marca_sf':marca_sf, 'Modelo_sf':modelo_sf, 'Año_sf':año_sf, 'millaje_sf':millaje_sf, 'precio_sf':precio_sf})
```
